##Visualizations:

```python
#stresslevel vs sleepquality

plt.figure(figsize=(10, 5))
plt.style.use('ggplot')
sns.violinplot(data=sleep, x='Stress Level', y='Quality of Sleep', color='#f363d9')
plt.title('Stress Level vs Sleep Quality')
plt.xlabel('Stress Level')
plt.ylabel('Quality of sleep')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=sleep, x='Stress Level', y='Sleep Duration')
plt.title('Stress Level vs Sleep Duration')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.kdeplot(sleep[sleep['Gender'] == 'Female']['Stress Level'], bw_adjust=0.8, fill=1, color='red')
sns.kdeplot(sleep[sleep['Gender'] == 'Male']['Stress Level'], bw_adjust=0.8, fill=1, color='green')
plt.title('Stress Level Distribution by Gender')
plt.xlabel('Stress Level')
plt.ylabel('Density')
plt.legend(['female', 'male'])
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.boxplot(data=sleep, x='Occupation', y='Stress Level', color = '#38c79b')
plt.title('Occupation and Stress Levels')
plt.xlabel('Occupation')
plt.ylabel('Stress Level')
plt.xticks(size=7)
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.boxplot(data=sleep, x='Occupation', y='Quality of Sleep', color='skyblue')
plt.title('Career vs. Quality of Sleep')
plt.xlabel('Career')
plt.ylabel('Quality of Sleep')
plt.xticks(size=7)
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.boxplot(data=sleep, y='Age', x='Quality of Sleep')
plt.title('Age vs. Quality of Sleep')
plt.xlabel('Quality of sleep')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.stripplot(data=sleep, x='Occupation', y='Quality of Sleep', hue='Gender')
plt.title('Quality of Sleep Distribution by Occupation and Gender')
plt.xlabel('Occupation')
plt.ylabel('Quality of Sleep')
plt.xticks(size=7)
plt.tight_layout()
plt.show()
```

```python
plt.scatter(sleep['Age'], sleep['Quality of Sleep'])
plt.xlabel('Age')
plt.ylabel('Quality of Sleep')
plt.title('Age vs. Quality of Sleep')
plt.show()
```

```python
plt.scatter(sleep['Age'], sleep['Stress Level'], color='blue')
plt.xlabel('Age')
plt.ylabel('Stress Level')
plt.title('Age vs. Stress Level')
plt.show()
```
