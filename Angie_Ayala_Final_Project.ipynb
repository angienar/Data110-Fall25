##First Visualization:

```python
plt.style.use('ggplot')
```

```python
traffic_violations['Date_Of_Stop_Month'] = pd.to_datetime(traffic_violations['Date Of Stop'], errors='coerce').dt.month

month_counts = traffic_violations['Date_Of_Stop_Month'].value_counts().sort_index()

plt.plot(month_counts.index, month_counts.values, marker='o', linestyle='-')
plt.title('Traffic Violations by Month')
plt.xlabel('Month')
plt.ylabel('Amount')

month_labels = ['January','February','March','April','May','June','July','August','September','October','November','December']
plt.xticks(range(1, 13), month_labels, rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

The first visualization I created was a line plot showing the amount of traffic violations that happened each month. The x-axis of this line plot is the month, and the y-axis is the amount. Looking at this graph I noticed that the month that has the most amount of violations March with more then 250,000 traffic violations. The month with the least number of violations is June, with fewer than 200,000.

##Second Visualization:

```python
traffic_violations['Time_of_Stop_Hour'] = pd.to_datetime(traffic_violations['Time Of Stop'], errors='coerce').dt.hour

plt.figure(figsize=(10, 5))
sns.kdeplot(traffic_violations[traffic_violations['Gender'] == 'F']['Time_of_Stop_Hour'].dropna(), bw_adjust=0.8, fill=0.5, color='red')
sns.kdeplot(traffic_violations[traffic_violations['Gender'] == 'M']['Time_of_Stop_Hour'].dropna(), bw_adjust=0.8, fill=0.5, color='green')
plt.title('Time of Stop Distribution by Gender')
plt.xlabel('Hour of Stop')
plt.ylabel('Density')
plt.legend(['Female', 'Male'])
plt.show()
```

The next visualization I created is a distribution graph, where the x-axis values are the time of the stop. The graph is divided by gender, with female represented by red and male by green. I noticed that males are stopped more between the hours of 0 to 4 and 20 to 23. Females were stopped more, 7 to 17. It seems that overall there are less stops in the morning and the most stops mid-day. There are multiple stops at the end of the day, but it is less than mid-day.

##Third Visualization:

```python
print(traffic_violations['Race'].value_counts())
```

```python
races_to_filter = ['BLACK', 'HISPANIC', 'OTHER', 'ASIAN', 'NATIVE AMERICAN','WHITE']
filtered_race_violations = traffic_violations[traffic_violations['Race'].isin(races_to_filter)]
print(f"Filtered violations count: {len(filtered_race_violations)}")
print(f"Unique races in filtered data:\n{filtered_race_violations['Race'].value_counts()}")
```

```python
race_counts = filtered_race_violations['Race'].value_counts()

plt.figure(figsize=(10, 6))
plt.barh(race_counts.index, race_counts.values, color='purple', edgecolor='black')
plt.xlabel('Count')
plt.ylabel('Race')
plt.title('Count of Traffic Violations by Race')
plt.show()
```

The visualization above is a bar graph showing the number of stops of each race. The race with the more amount of stop is black with 11,006, and the second was white with 9,931. The race with the least amount of stops is Native Americans with only 20.

##Fourth Visualization:

```python
traffic_violations['Latitude'] = pd.to_numeric(traffic_violations['Latitude'], errors='coerce')
traffic_violations['Longitude'] = pd.to_numeric(traffic_violations['Longitude'], errors='coerce')

cleaned_violations = traffic_violations.dropna(subset=['Latitude', 'Longitude'])

sample_violations = cleaned_violations.sample(n=1000, random_state=42)
```

```python
import folium

# Calculate the center of the map using the mean latitude and longitude of the sampled data
map_center_latitude = sample_violations['Latitude'].mean()
map_center_longitude = sample_violations['Longitude'].mean()

# Create a Folium map centered around the average coordinates
m = folium.Map(location=[map_center_latitude, map_center_longitude], zoom_start=5)

# Add markers for each violation in the sampled data
for index, row in sample_violations.iterrows():
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=1,  # Small radius for individual points
        color='red',
        fill=True,
        fill_color='red',
        fill_opacity=0.6
    ).add_to(m)

# Display the map
display(m)
```
