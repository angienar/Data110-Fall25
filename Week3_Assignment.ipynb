##Adjusting the Plot: Line Colors and Styles

```python
plt.plot(x, np.sin(x - 0), color='blue') # specify color by name
plt.plot(x, np.sin(x - 1), color='g') # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75') # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44') # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # HTML color names supported
```

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

#For short, you can use the folowwing codes:
plt.plot(x, x +4, linestyle='-') # solid
plt.plot(x, x +5, linestyle='--') # dashed
plt.plot(x, x +6, linestyle='-.') # dashdot
plt.plot(x, x +7, linestyle=':'); # dotted
```

```python
plt.plot(x, x + 0, '-g') # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r'); # dotted red
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend();
```

##8. Customizing Scatter Plots

- Generate data for the function $f(x)= xsin(x)$ over the range [0,4]
- Create a scatter plot of this data with a line connecting the points
- Customize the plot by changing the marker size, line width, marker face color, marker edge color, and marker edge width

```python
x= np.linspace(0, 4*np.pi, 20)
y= x*np.sin(x)

plt.plot(x, y, '-p', color='g', markersize=10, linewidth=4, markerfacecolor='white', markeredgewidth=2, label='$f(x)= xsin(x)$')
plt.title('Customizing Scatter Plots')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower left')
plt.show()
```

##9. Combined Plot

- Plot the function $f(x)=xcos(x)$ over the range [-2, 2]
- Add scatter points for x values at every $\frac{}{4}$ interval
- Use different colors and styles for the line and scatter components

```python
x1= np.linspace(-2*np.pi, 2*np.pi,100)
x2= np.linspace(-2*np.pi, 2*np.pi,4) # Changed to 4 points so it only includes every pi/4 interval
y1 = x1*np.cos(x1)
y2 = x2*np.cos(x2)


plt.plot(x1, y1, color='r', linestyle='--', label='f(x)')
plt.scatter(x2, y2, marker='o', color='b', label='scatter points')
plt.title('Combined Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
