###Step 2: Anscombe's Quartet Dataset

```python
# Anscombe's Quartet:
x = [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5]
y1 = [8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68]
y2 = [9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74]
y3 = [7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73]
x4 = [8, 8, 8, 8, 8, 8, 8, 19, 8, 8, 8]
y4 = [6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89]
```

```python
plt.scatter(x, y1)
```

```python
# Calculate the slope (m) and intercept (b) of the line using np.polyfit
m, b = np.polyfit(x, y1, 1)

# Create the regression line
regression_line = m * np.array(x) + b

# Plot the data points and regression line
plt.scatter(x, y1)
plt.plot(x, regression_line,color='red')
plt.xlabel('x')
plt.ylabel('y1')
```

###What is your reflection?

**How do the datasets visually differ despite having similar summary statistics?**

Although the datasets have similar summary statistics, the points on the graph are different. For example, the x and y3 graph shows points in the shape of an increasing line with one outlier, while the x and y4 graphs' points are scattered. Also, all the regression lines are noticeably different.

**How did the outlier in the x4, y4 dataset affect the regression line compared to the other datasets?**

The outlier in x4 and y4 affects the regression line because, without the outlier, the regression line would have a steeper slope than it does and look very different.


**Why is it important to visualize data in addition to calculating summary statistics?**

It is important to visualize data and calculate summary statistics because we are able to see all the points location and if they make a shape, are scattered around, or if there are any outliers. It is also important because it condenses the data and allows us to identify patterns.
