#**Penguin Dataset**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Load the penguins dataset from seaborn
penguins = sns.load_dataset('penguins')

# Display the first few rows of the dataset
print(penguins.head())
```

```python

penguins.info()
```

```python
penguins.describe()
```

```python
plt.scatter(penguins['bill_length_mm'], penguins['bill_depth_mm'])
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.title('Bill Length vs. Bill Depth')
plt.show()
```

```python
# Set figure size
plt.figure(figsize=(8, 5))

# Scatter plot (manually colored)
plt.scatter(penguins["bill_length_mm"], penguins["bill_depth_mm"], color="dodgerblue", alpha=0.6, edgecolors="black", label="Data Points")

# Regression line
sns.regplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", scatter=False, color="red", ci=None, label="Regression Line")

# Labels and title
plt.xlabel("Bill Length (mm)", fontsize=12)
plt.ylabel("Bill Depth (mm)", fontsize=12)
plt.title("Penguin Bill Measurements with Regression", fontsize=14)

# Add legend
plt.legend()

# Show plot
plt.show()
```

##**Task 5: Create a 22 Grid with Mixed Plot Types**

Make a figure with **four subplots** arranged in a 22 grid using
fig, axes = plt.subplots(2, 2, figsize=(10, 8)).

Use the **penguins dataset** and include a mix of plot types:



*   **(Top-left):** sns.barplot()  show **body mass by species.**
*   **(Top-right):** sns.scatterplot()  plot **bill length vs bill depth**, color by species.
*  **(Bottom-left):** sns.histplot()  distribution of **flipper length**, separated by species (hue="species").
*   **(Bottom-right):** sns.boxplot()  compare **body mass by island.**

Add a **title for each subplot**, and call plt.tight_layout() at the end.

**Deliverable:** one combined figure (saved as .png, around 150 dpi) that clearly shows all four plots with appropriate labels and colors.

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

#subplot 1: Top-left - sns.barplot() for body mass by species
sns.barplot(data=penguins, x='species', y='body_mass_g', color = '#f23943',ax=axes[0, 0])
axes[0, 0].set_title('Body Mass by Species')
axes[0, 0].set_xlabel('Species')
axes[0, 0].set_ylabel('Body Mass (g)')

#subplot 2: Top-right - sns.scatterplot() for bill length vs bill depth, color by species
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species', palette='viridis', ax=axes[0, 1])
axes[0, 1].set_title('Bill Length vs Bill Depth')
axes[0, 1].set_xlabel('Bill Length (mm)')
axes[0, 1].set_ylabel('Bill Depth (mm)')
axes[0, 1].legend(title='Species', loc='upper right', bbox_to_anchor=(1.2, 1.05)) #bbox_to_anchor moves the legend up and down or left and right to move it outside the graph

#subplot 3: Bottom-left - sns.histplot() for flipper length distribution, separated by species
sns.histplot(data=penguins, x='flipper_length_mm', hue='species', kde=True, palette='magma', ax=axes[1, 0])
axes[1, 0].set_title('Flipper Length Distribution')
axes[1, 0].set_xlabel('Flipper Length (mm)')
axes[1, 0].set_ylabel('Density')

#subplot 4: Bottom-right - sns.boxplot() for body mass by island
sns.boxplot(data=penguins, x='island', y='body_mass_g', color='#5961e5',ax=axes[1, 1])
axes[1, 1].set_title('Body Mass by Island')
axes[1, 1].set_xlabel('Island')
axes[1, 1].set_ylabel('Body Mass (g)')

plt.tight_layout()
plt.show()
```

##**Task 6: Create a Pie Chart (Penguins Dataset)**

1.Count how many penguins belong to each species using value_counts().

2.Create a pie chart using plt.pie() with:

  * labels as species names.
  * autopct='%1.1f%%' to show percentages.
  * startangle=90 for better orientation.
  * explode to slightly separate one slice (optional).

3.Add a title like "Distribution of Penguin Species".

4.Make the chart circular by setting plt.axis('equal').

```python
species_counts = penguins['species'].value_counts()
```

```python
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1d%%', startangle=90, colors=['lightblue', 'pink', 'lavender'])
plt.title('Distribution of Penguin Species')
plt.axis('equal')
plt.show()
```
